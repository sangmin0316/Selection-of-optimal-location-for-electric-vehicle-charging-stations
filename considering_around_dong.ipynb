```python
# 주변 동에 평균 충전기 구할때 필요없는 colums을 삭제하는 함수를 만듭시다.
def drop_col_df(df):
  df = df.drop('Unnamed: 0', axis = 1)
  df = df.drop('dong', axis = 1)
  df = df.drop('X', axis = 1)
  df = df.drop('Y', axis = 1)
  df = df.drop('highway', axis = 1)
  df = df.drop('총생활인구수', axis = 1)
  return df
```

```python
# concat을 이용할건데, 먼저 0번째 행에대해 df를 만듭니다.
# 그러면 concat을 할때 처음 시작 점(기준?)을 잡을 수 있습니다.
# a = concat([a,b]) 이때 a를 만드는 작업입니다.
def make_first_df(df):
  DCdf = drop_col_df(df)
  row_num = [] # j를 담을 list
  distance = [] # 거리 값을 담을 list 
  x1 = df.loc[0,'X']
  y1 = df.loc[0,'Y']
  for j in range (0,len(df)):
    x2 = df.loc[j,'X']
    y2 = df.loc[j,'Y']
    r = ((x1-x2)**2 + (y1-y2)**2) 
    row_num.append(j)
    distance.append(r)
    if r == 0: # r = 0 즉 자기 자신이 나오면 r을 큰값으로 둬서 저절로 삭제하게 만듭시다.
      r = 10
    if len(row_num) == 6:
      del row_num[distance.index(max(distance))]
      distance.remove(max(distance))
    else:
      pass     
  total_index = df.index
  reverse_index = list(set(total_index) - set(row_num))     
  around_dong = DCdf.drop(reverse_index)
  around_dong = around_dong.apply(np.average, axis = 0)
  data = np.reshape(around_dong.values,(1,8))
  ADdf = pd.DataFrame(data = data, columns = ['affnc',	'affc',	'anfnc',	'anfc',	'afsnc',	'afsc',	'ansnc',	'ansc'])
  a = df.loc[0,'dong'] # df1에서 i번째 dong명을 가져옵니다, 
  ADdf['dong'] = a    
  return ADdf
```

```python
def loc(df):
  DCdf = drop_col_df(df)
  around_df = make_first_df(df)
  for i in range (1,len(df)):
    row_num = [] # j를 담을 list
    distance = [] # 거리 값을 담을 list 
    x1 = df.loc[i,'X']
    y1 = df.loc[i,'Y']
    for j in range (0,len(df)):
      x2 = df.loc[j,'X']
      y2 = df.loc[j,'Y']
      r = ((x1-x2)**2 + (y1-y2)**2) 
      row_num.append(j)
      distance.append(r)
      if r == 0: # r = 0 즉 자기 자신이 나오면 r을 큰값으로 둬서 저절로 삭제하게 만듭시다.
        r = 10
      if len(row_num) == 6:
        del row_num[distance.index(max(distance))]
        distance.remove(max(distance))
      else:
        pass     
    total_index = df.index
    reverse_index = list(set(total_index) - set(row_num))     
    around_dong = DCdf.drop(reverse_index)
    around_dong = around_dong.apply(np.average, axis = 0)
    # print(around_dong)
    data = np.reshape(around_dong.values,(1,8))
    ADdf = pd.DataFrame(data = data, columns = ['affnc',	'affc',	'anfnc',	'anfc',	'afsnc',	'afsc',	'ansnc',	'ansc'])
    a = df1.loc[i,'dong'] # df1에서 i번째 dong명을 가져옵니다, 
    ADdf['dong'] = a    
    # print(ADdf)
    around_df = pd.concat([around_df,ADdf])
  return around_df
```
