```python
import pandas as pd

HWdf = pd.read_excel('고속도로.xlsx')
df1 = pd.read_excel('029df.xlsx')
df2 = pd.read_excel('9218df.xlsx')
df3 = pd.read_excel('18224df.xlsx')
```

```python
# drop_duplicates()를 이용해서 중복되는것을 제거해줍시다. 
HWdf = HWdf.drop_duplicates()
```

```python
HWdf.info()
```

```python
df1.info()
```

```python
# HWdf에 있는 동이 df1,2,3에 다 매칭이 되는지 확인해보겠습니다.
a = 0
for i in range(0,len(HWdf)):
  b = HWdf.iloc[i,0] # 행정동코드의 행정동명
  if len(df1[df1['dong'] == str(b)]) == 1: # 일치하는게 1개 있다면~
    a = a + 1
  elif len(df1[df1['dong'] == str(b)]) >= 2: # 만약 중복되는게 있다면~
    print('overlap:', b)
  else:
    print(b) # 하나도 없다면~
print(a) # 행정동코드의 동과 tdf1에서의 동이 일치하는 갯수
```

```python
# 고속도로에 있는 데이터에서 필요한 데이터인 weight와 dong만 남깁시다.
HWdf = HWdf.drop('고속도로이름', axis = 1)
HWdf = HWdf.drop('구', axis = 1)
```

```python
HWdf.head()
# 저희가 필요한데이터만 남았네요!
```

```python
# weight의 이름을 바꿔 줍시다.
HWdf = HWdf.rename(columns = {'weight' : 'highway'})
```

```python
#이제 데이터들을 합쳐봅시다.
df1 = pd.merge(HWdf, df1, how = 'right')
```

```python
df1.head(30)
```

```python
df1.info()
# 고속도로 가중치 27개가 잘 들어갔네요.
# dong이 424개여서 427개로 늘어났습니다. 
# 위에 head()를보면 고덕1동 고덕 2동이 중복되어있는 것을 확인할 수있습니다.
```

```python
#중복을제거해주고
#마지막으로 NaN값을 0으로 채워줍시다.
df1 = df1.drop_duplicates()
df1 = df1.fillna(0)
```

```python
df1.info()
```

```python
# df2, df3에도 위와 같은 작업을 진행해줍니다.
df2 = pd.merge(HWdf, df2, how = 'right')
df2 = df2.drop_duplicates()
df2.fillna(0)
df3 = pd.merge(HWdf, df3, how = 'right')
df3 = df3.drop_duplicates()
df3.fillna(0)
```

```python
# 만든 파일을 저장해둡시다.
df1.to_excel('029_df.xlsx', index = False)
df2.to_excel('9218_df.xlsx',index = False)
df3.to_excel('18224_df.xlsx',index = False)
```
