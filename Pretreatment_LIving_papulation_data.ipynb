```python
# 생활인구 데이터를 가공하려고 합니다.
# 생활인구 데이터의 행정동을 살펴보고자 하는데, 행정동 명칭으로 주어지지 않고 행정동코드로 주어져있습니다.
# 행정동코드가 각 어떤 행정동을 나타내는지는 행정동코드_매칭정보 파일에 따로 지정되어있습니다.
# 첫째로 우리가 할 일은 생활인구 데이터의 행정동코드를 행정동이름으로 바꾸는 것입니다.

import pandas as pd
df1 = pd.read_excel('LIVING_POPULATION.xlsx')
```

```python
df1.head()
```

```python
# 날짜에 해당하는 열이 Unnamed: 0으로 나타나고 있습니다. rename을 이용해서 바꿔줍니다. 
df1 = df1.rename(columns={'Unnamed: 0' : '날짜'})
```

```python
df2 = pd.read_excel('행정동코드_매칭정보.xlsx')
```

```python
df2.head()
```

```python
# 우리가 사용할 행정동코드는 행자부행정동코드로 나와있습니다. 이것을 행정동코드입니다.
df2 = df2.rename(columns={'행자부행정동코드' : '행정동코드'})
df2.head()
```

```python
# 행정동코드 열만 따라 가져와서 DataFrame을 만들어줍니다.
# 여러 시도 끝에 작성된 코드가 행정동 column만 바꾸지 못하고 전체 column을 바꾸는 문제가 있었습니다.
# 그래서 행정동코드 열을 따로 가져와서 처리하려고합니다.
tdf = df1['행정동코드']
ttdf = pd.DataFrame(tdf)
ttdf.head()
```

```python
# 이것이 여러 시도 끝에 작성한 코드입니다..
# 원래는 ttdf대신 df1을 이용했는데 column전체가 b로 바뀌는 문제가 있었습니다.
for i in range(1,len(df2)):
  a = df2['행정동코드'][i]
  b = df2['행정동명'][i]
  ttdf[(ttdf['행정동코드'] == int(a))] = b
ttdf.head(10)
```

```python
# 행정동코드 열에 행정동 명이 들어간것을 확인할 수 있습니다
df1['행정동코드'] = ttdf['행정동코드']
df1.head()
```

```python
# 행정동코드 열의 이름을 행정동명으로 바꿔 줍시다.
df1 = df1.rename(columns={'행정동코드':'행정동명'})
```

```python
df1.head()
```

```python
# 원하는 결과가 나왔으니 파일을 저장해줍시다.
df1.to_excel('code_to_name_LPdf.xlsx')
```
