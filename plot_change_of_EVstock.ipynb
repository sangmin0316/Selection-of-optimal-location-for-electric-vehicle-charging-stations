```python
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
x = np.arange(2017,2022)
print(x)
```

```python
# 데이터를 na.array로 가져왔습니다.
fasty = np.array([3343, 5213, 7396, 9805, 12789])
slowy = np.array([10333, 22139, 37396, 54383, 59316])
totaly = fasty + slowy
```

```python
# 급속충전소 변화량과  25년 목표치
plt.scatter(x,fasty)
plt.plot(x,fasty,'--', label ='The numeber of fast charging EV stock')
plt.scatter([2025],[12000])
plt.xlabel('Time(year)')
plt.ylabel('Numeber')
plt.legend()
# 자료가 이상한것 같음, 2025년 목표치가 1.2만개라고 자료에 나오는데 2021년 급속충전기 갯수 12789개..
```

```python
# 총 충전기량 변화수를 그래프로 그리는거
plt.bar(x,totaly, label ='The numeber of EV stock')
plt.bar([2025],[620000], label = 'The number of goal EV stock')
plt.xlabel('Time(year)')
plt.ylabel('Numeber')
plt.title('Changes in the number of EV stock in Korea.')

# 2022 ~ 2024년도는 그래프 상에 표시 하지 않으려고 xticks를 이용했습니다.
plt.xticks(np.arange(2017,2026), labels=['2017','2018','2019','2020','2021','','','','2025'])

#plt.text로 그래프이 값 표시
plt.text(2025,620000,
          '62000',
          color='black',
          horizontalalignment='center',
          verticalalignment='bottom')

for i in range(len(x)):
  plt.text(x[i],totaly[i],
          str(totaly[i]),
          color='black',
          horizontalalignment='center',
          verticalalignment='bottom')
plt.legend()
plt.savefig('Change_EVstock_korea.png',bbox_inches='tight', pad_inches=1)
```
